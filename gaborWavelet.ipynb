```python
def build_filters():
    filters = []
    ksize = 16
    for theta in np.arange(0, np.pi, np.pi / 15):
        kern = cv2.getGaborKernel((ksize, ksize), 4.0, theta, 10.0, 0.5, 0, ktype=cv2.CV_32F)
        kern /= 1.5*kern.sum()
        filters.append(kern)
        print(theta)
    return filters
```

```python
filters = build_filters()
```

```python
def process(img, filters):
    accum = []
    for kern in filters:
        fimg = cv2.filter2D(img, cv2.CV_8UC3, kern)
        temp = 0
        for i in fimg:
            for j in i:
                temp += j
        accum.append(temp)
    return accum
```

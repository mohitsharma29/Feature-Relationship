```python
winSize = (128,128)
blockSize = (32,32)
blockStride = (16,16)
cellSize = (16,16)
nbins = 18
derivAperture = 1
winSigma = -1.
histogramNormType = 0
L2HysThreshold = 0.2
gammaCorrection = 1
nlevels = 64
useSignedGradients = True
```

```python
hog = cv2.HOGDescriptor(winSize,blockSize,blockStride,cellSize,nbins,derivAperture,winSigma,histogramNormType,L2HysThreshold,gammaCorrection,nlevels, useSignedGradients)
```

```python
data = scipy.io.loadmat('lists_var_size.mat')['list']
```

```python
dataDict = {}
dataDict['labels'] = data[0][0][0]
dataDict['names'] = data[0][0][1]
dataDict['classlabels'] = data[0][0][2]
dataDict['classnames'] = data[0][0][3]
dataDict['numclasses'] = data[0][0][4][0]
dataDict['trnInd'] = data[0][0][5]
dataDict['tstInd'] = data[0][0][6]
dataDict['valInd'] = data[0][0][7]
dataDict['txnInd'] = data[0][0][8]
```

```python
numExam = 1016
numClasses = 62
```

```python
hogVectors = []
```

```python
k = 0
for i in range(numClasses):
    temp_vecs = []
    for j in range(numExam):
        temp_img = cv2.imread('Fnt/' + dataDict['names'][k] + '.png', cv2.IMREAD_GRAYSCALE)
        k += 1
        #print(dataDict['names'][k])
        temp_H = hog.compute(temp_img).tolist()
        temp_vecs.append(temp_H)
    #print(len(temp_vecs), len(temp_vecs[0]))
    final_vec = []
    for j in range(len(temp_vecs[0])):
        temp_sum = 0
        for l in range(numExam):
            temp_sum += temp_vecs[l][j][0]
        final_vec.append(temp_sum/numExam)
    hogVectors.append(final_vec)
    print("Done Class " + str(i))
```
